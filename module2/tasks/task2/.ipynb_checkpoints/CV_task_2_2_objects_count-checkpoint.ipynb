```python
import scipy as sc
import cv2
import numpy as np
import imutils
from skimage import io, feature, morphology 
from skimage.feature import peak_local_max
from skimage.morphology import watershed
from scipy import ndimage
from matplotlib import pyplot as plt
#from google.colab.patches import cv2_imshow

def url_to_image(url):
  print("downloading %s" % (url))
  return cv2.cvtColor(io.imread(url), cv2.COLOR_BGR2RGB)
```

### count1.jpg

```python
image = cv2.imread('../../images/segmentaion/count/count1.jpg')
#url1 = "https://raw.githubusercontent.com/mdavydov/ComputerVisionCourse/master/tasks/count/count1.jpg"
#image = url_to_image(url1)

#apply Gaussian smooting, pyrMeanShiftFiltering, grey scale and adaptive threshold
gaussian = cv2.GaussianBlur(image,(5,5),0)
shifted = cv2.pyrMeanShiftFiltering(gaussian, 41, 51)
gray = cv2.cvtColor(shifted, cv2.COLOR_BGR2GRAY)
thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY, 11, 5)
 
D = ndimage.distance_transform_edt(thresh)
localMax = peak_local_max(D, indices=False, min_distance=5,  
    labels=thresh)
# perform a connected component analysis on the local peaks,
# using 8-connectivity, then apply the Watershed algorithm
markers = ndimage.label(localMax, structure=np.ones((3, 3)))[0]
labels = watershed(-D, markers, mask=thresh)
print("Number of unique segments automatically detected: ",(len(np.unique(labels)) - 1))

# loop over the unique labels returned by the Watershed
# algorithm
r_list = []
for label in np.unique(labels):
    # if the label is zero, we are examining the 'background'
    # so simply ignore it
    if label == 0:
        continue
 
    # otherwise, allocate memory for the label region and draw
    # it on the mask
    mask = np.zeros(gray.shape, dtype="uint8")
    mask[labels == label] = 255
 
    # detect contours in the mask and grab the largest one
    cnts = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cnts = imutils.grab_contours(cnts)
    c = max(cnts, key=cv2.contourArea)
 
    # draw a circle enclosing the object
    ((x, y), r) = cv2.minEnclosingCircle(c)
    r_list.append(r)
    
    if r <15: #threshold value based on histogram below
        cv2.circle(image, (int(x), int(y)), int(r), (0, 255, 0), 1)

plt.figure(1)
plt.imshow(image)
plt.title("count1.jpg")
plt.figure(2)
plt.imshow(thresh, cmap = 'gray')
```

```python
plt.hist(r_list,40,[0,40])
plt.show()

threshold = 16
k=0
for i in r_list:
    if i < threshold:
        k+=1
print("Number of elements: ", k)
print("error: ",round((1-k/len(r_list)),2)*100,"%")
```

### count2.jpg

```python
img = cv2.imread('../../images/segmentaion/count/count2.jpg')

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gaussian = cv2.GaussianBlur(gray,(5,5),0)

ret, thresh = cv2.threshold(gaussian,80,255,cv2.THRESH_BINARY)

D = ndimage.distance_transform_edt(thresh)
localMax = peak_local_max(D, indices=False, min_distance=10,  
    labels=thresh)
# perform a connected component analysis on the local peaks,
# using 8-connectivity, then apply the Watershed algorithm
markers = ndimage.label(localMax, structure=np.ones((3, 3)))[0]
labels = watershed(-D, markers, mask=thresh)
print("Number of unique segments automatically detected: ",(len(np.unique(labels)) - 1))

# loop over the unique labels returned by the Watershed
# algorithm  
for label in np.unique(labels):
    # if the label is zero, we are facing with the 'background'
    if label == 0:
        continue
    # otherwise, allocate memory for the label region and draw
    # it on the mask
    mask = np.zeros(gray.shape, dtype="uint8")
    mask[labels == label] = 255
    # detect contours in the mask
    cnts = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL,
        cv2.CHAIN_APPROX_SIMPLE)[-2]
    c = max(cnts, key=cv2.contourArea)
    rect = cv2.minAreaRect(c)
    box = cv2.boxPoints(rect)
    box = np.int0(box)  
    cv2.drawContours(img,[box],-1,(255,0,0),3)  

plt.figure(1)
plt.imshow(img)
plt.title("count2.jpg")
plt.figure(2)
plt.imshow(thresh, cmap = 'gray')
```

### count8.jpg

```python
img = cv2.imread('../../images/segmentaion/count/count8.jpg')

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gaussian = cv2.GaussianBlur(gray,(5,5),0)

ret, thresh = cv2.threshold(gaussian,145,255,cv2.THRESH_BINARY)

D = ndimage.distance_transform_edt(thresh)
localMax = peak_local_max(D, indices=False, min_distance=15,  
    labels=thresh)
# perform a connected component analysis on the local peaks,
# using 8-connectivity, then apply the Watershed algorithm
markers = ndimage.label(localMax, structure=np.ones((3, 3)))[0]
labels = watershed(-D, markers, mask=thresh)
print("Number of unique segments defined automatically: ",(len(np.unique(labels)) - 1))

# loop over the unique labels returned by the Watershed
# algorithm  
size_list = []
for label in np.unique(labels):
    # if the label is zero, we are facing with the 'background'
    if label == 0:
        continue
    # otherwise, allocate memory for the label region and draw
    # it on the mask
    mask = np.zeros(gray.shape, dtype="uint8")
    mask[labels == label] = 255
    # detect contours in the mask
    cnts = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL,
        cv2.CHAIN_APPROX_SIMPLE)[-2]
    c = max(cnts, key=cv2.contourArea)
    rect = cv2.minAreaRect(c)
    box = cv2.boxPoints(rect)
    box = np.int0(box)  
    size_list.append(cv2.contourArea(c))
    if 600<cv2.contourArea(c)<1800: #min and max thresholds checked by visualizing outliers
        cv2.drawContours(img,[box],-1,(255,0,0),2)  

plt.figure(1)
plt.imshow(img)
plt.title("count8.jpg")
plt.figure(2)
plt.imshow(thresh, cmap = 'gray')
```

```python
D = ndimage.distance_transform_edt(thresh)
localMax = peak_local_max(D, indices=False, min_distance=20,  
    labels=thresh)
# perform a connected component analysis on the local peaks,
# using 8-connectivity, then apply the Watershed algorithm
markers = ndimage.label(localMax, structure=np.ones((3, 3)))[0]
labels = watershed(-D, markers, mask=thresh)
print("Number of unique segments defined automatically: ",(len(np.unique(labels)) - 1))

# loop over the unique labels returned by the Watershed
# algorithm  
size_list = []
for label in np.unique(labels):
    # if the label is zero, we are facing with the 'background'
    if label == 0:
        continue
    # otherwise, allocate memory for the label region and draw
    # it on the mask
    mask = np.zeros(thresh.shape, dtype="uint8")
    mask[labels == label] = 255
    # detect contours in the mask
    cnts = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL,
        cv2.CHAIN_APPROX_SIMPLE)[-2]
    c = max(cnts, key=cv2.contourArea)
    rect = cv2.minAreaRect(c)
    box = cv2.boxPoints(rect)
    box = np.int0(box)  
    size_list.append(cv2.contourArea(c))
    cv2.drawContours(result,[box],-1,(255,0,0),2)  

plt.figure(4)
plt.imshow(result)
plt.title("count11.jpg")
```

### count12.jpg

```python
img = cv2.imread('../../images/segmentaion/count/count12.jpg')

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY, 15, 5)

D = ndimage.distance_transform_edt(thresh)
localMax = peak_local_max(D, indices=False, min_distance=10,  
    labels=thresh)
# perform a connected component analysis on the local peaks,
# using 8-connectivity, then apply the Watershed algorithm
markers = ndimage.label(localMax, structure=np.ones((3, 3)))[0]
labels = watershed(-D, markers, mask=thresh)
print("Number of unique segments detected automatically: ",(len(np.unique(labels)) - 1))

# loop over the unique labels returned by the Watershed
# algorithm
r_list = []
for label in np.unique(labels):
    # if the label is zero, we are examining the 'background'
    # so simply ignore it
    if label == 0:
        continue
 
    # otherwise, allocate memory for the label region and draw
    # it on the mask
    mask = np.zeros(gray.shape, dtype="uint8")
    mask[labels == label] = 255
 
    # detect contours in the mask and grab the largest one
    cnts = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cnts = imutils.grab_contours(cnts)
    c = max(cnts, key=cv2.contourArea)
 
    # draw a circle enclosing the object
    ((x, y), r) = cv2.minEnclosingCircle(c)
    r_list.append(r)
    #draw circles
    if 4<r<16:
        cv2.circle(img, (int(x), int(y)), int(r), ( 255, 0,0), 2)

plt.figure(1)
plt.imshow(img)
plt.title("coun12.jpg")
plt.figure(2)
plt.imshow(thresh, cmap = 'gray')
```
