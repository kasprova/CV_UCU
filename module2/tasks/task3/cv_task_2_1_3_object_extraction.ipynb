```python
import scipy as sc
import cv2
import numpy as np
import imutils
from skimage import io, feature, morphology 
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

def url_to_image(url):
  print("downloading %s" % (url))
  return cv2.cvtColor(io.imread(url), cv2.COLOR_BGR2RGB)
```

```python
def extract_object(image, rectangle, n_iterations = 5):
  """
  Function to extract pointed out objects from background using grabcut algorithm
  Input: 
    image - input image: numpy array
    rectangle - coordinates of a rectangle which includes the foreground object in the format (x,y,w,h)
    n_iterations - number of iterations the algorithm should run.
  Output: 
    the image of the object with cut background: numpy array
  
  """
  
  #initialize mask image where we specify which areas are background, foreground etc.
  mask = np.zeros(image.shape[:2],np.uint8)
  
  #internal arrays
  bgdModel = np.zeros((1,65),np.float64)
  fgdModel = np.zeros((1,65),np.float64)
  
  #apply grabcut algorithm
  cv2.grabCut(img = image, mask = mask,rect = rectangle, bgdModel = bgdModel, fgdModel = fgdModel, iterCount = n_iterations, mode = cv2.GC_INIT_WITH_RECT)
  
  mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
  
  image = image*mask2[:,:,np.newaxis]
  
  return image
  
```

### obj1.jpg

```python
#read image
url = "https://raw.githubusercontent.com/kasprova/CV_UCU/master/module2/images/segmentaion/object/obj1.jpg"
img = url_to_image(url)
#cv2_imshow(img)

plt.imshow(img),plt.colorbar(),plt.show()
```

```python
output =  extract_object(img, (40,20,410,230), 9)
cv2_imshow(output)
```
