### Task 2.4: Given 3 pairs of points find affine transformation and apply it an image.

```python
import numpy as np
import cv2
from skimage import io
from matplotlib import pyplot as plt
from numpy.linalg import inv
```

```python
img = cv2.imread('../../images/correspondence/view.jpg')
#Affine transformations:
#x'=a11*x+a12*y+b1
#y'=a21*x+a22*y+b2

(r,c) = img.shape[:2]
pts1 = np.float32([[0,0],[r,0],[0,c]])
pts2 = np.float32([[50,0],[r,50],[0,c-50]])


M = cv2.getAffineTransform(pts1,pts2)

# implement function - cv2.getAffineTransform(pts1,pts2) using numpy
def my_getAffineTransform(pts1,pts2):
  """
  Equation of Affine transformations:
  |x0' x1' x2'| = |a11 a12 b1| |x0 x1 x2|
  |y0' y1' y2'|   |a21 a22 b2| |y0 y1 y2|
                               | 1  1  1|
  
  Find: M = |a11 a12 b1|
            |a21 a22 b2|
  
  M=X'Xinv
                                      
  Input:
    pts1 - 2D array of initial 3 points having the following structure:
        np.array([[x0,y0],[x1,y1],[x2,y2]])
    pts2 - 2D array of result 3 points having the following structure:
        np.array([[x'0,y'0],[x'1,y'1],[x'2,y'2]])
  Output:
    2D array of coefficients, having the followign structure
        np.array([[a11,a12,b1],[a21,a22,b2])
  
  """
  
  def build_matrix_of_initial_values(input):
    """
    Input: 
      input - 2D array of initial 3 points having the following structure:
        np.array([[x0,y0],[x1,y1],[x2,y2]])
    Output: 2D array having the structure
        np.array([[x0,x1,x2],[y0,y1,y2],[1,1,1]])
        
        build |x0 x1 x2|
              |y0 y1 y2|
              | 1  1  1|
    """
    h, w = pts1.shape[::-1]
    h+=1
    A = np.zeros((h,w), np.int32)
    for i, row in enumerate(input):
      for j, value in enumerate(row):
        A[j,i] = value
      A[h-1,i] = 1 
    return A
  
  def build_matrix_of_result_values(input):
    """
    Input:
      input - 2D array of result 3 points having the following structure:
        np.array([[x'0,y'0],[x'1,y'1],[x'2,y'2]])
    Output: 2D array having the structure
        np.array([[x'0,x'1,x'2],[y'0,y'1,y'2]])
        
        build |x0' x1' x2'|
              |y0' y1' y2'|        
    
    """
    h, w = pts1.shape[::-1]
    A = np.zeros((h,w), np.int32)
    for i, row in enumerate(input):
      for j, value in enumerate(row):
        A[j,i] = value
    return A
  
  #initialization
  X = build_matrix_of_initial_values(pts1)
  X_ = build_matrix_of_result_values(pts2)
  
  return np.dot(X_,inv(X))  

M_my = my_getAffineTransform(pts1,pts2)


#compare results of CV2 build in function and function written by me:)
print("CV2 M: ",M)
print("My M: ", M_my)

dst = cv2.warpAffine(img,M_my,img.shape[:2])

plt.figure(1,figsize=(20, 10))
plt.imshow(img)

plt.figure(2,figsize=(20, 10))
plt.imshow(dst)
plt.show()
```
